```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import accuracy_score
import pandas as pd
from rich.console import Console    
from sklearn.preprocessing import StandardScaler

# Dataframe Creation
movies = pd.DataFrame(marvel)
movies=movies.set_index('Title',drop=True)
movies=movies.sort_values(by='Box Office',ascending=False)
movies
```

```python
# Label Encoding Studio and Stars (into numeric)
label_encoder=LabelEncoder()
movies['Star Encoded'] = label_encoder.fit_transform(movies['Star'])
movies['Studio Encoded'] = label_encoder.fit_transform(movies['Studio'])
movies['Actors Encoded'] = label_encoder.fit_transform(movies['Actors'])

# Standardize numerical features
scaler = StandardScaler()
movies_scaled=movies.copy()
movies_scaled[['Year','Box Office','Phase']]=scaler.fit_transform(
    movies_scaled[['Year','Box Office','Phase']])

# Weighting features
movies_scaled[['Star Encoded']]=movies_scaled[['Star Encoded']]*100
movies_scaled[['Actors Encoded']]=movies_scaled[['Actors Encoded']]*100
movies_scaled[['Studio Encoded']]=movies_scaled[['Studio Encoded']]*1000
movies_scaled[['Year']]=movies_scaled[['Year']]
movies_scaled[['Box Office']]=movies_scaled[['Box Office']]
# Moving standardized values to features dataframe
features = movies_scaled[['Actors Encoded','Year','Star Encoded','Box Office','Studio Encoded']]
```

```python
# Start up Nearest Neighbors Model
model = NearestNeighbors(n_neighbors=7, metric='euclidean')
model.fit(features)

# Function for similar movies
console=Console()
def get_similar_movies(title):

    if title not in movies.index:
        print(f"Movie '{title}' not found in dataset.")
        return
    distances, indices = model.kneighbors(features.loc[[title]])
    similar_movies = movies.iloc[indices[0]]
    console.print('Movie Recommendations',style="bold")
    print('\n')
    return similar_movies[['Year','Actors','Box Office']]
```
